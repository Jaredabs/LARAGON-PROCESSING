```python
import matplotlib.pyplot as plt 
```

```python
x = [10,50,20,80]
y = [10,20,30,40]
plt.xlabel('Ejex')#texto en eje x
plt.ylabel('Ejey')#texto o titulo en y
plt.title('Ejemplo grafico de lineas')
plt.plot(x,y)#grafica de lineas en 2d
plt.show()
```

```python
plt.xlabel('Ejex')
plt.ylabel('Ejey')
plt.title('Ejemplo grafico de lineas')

```

```python
x1 = ['A','B','C','D']#Listas
y1 = [100,50,80,40]
plt.bar(x1,y1)#baras
plt.xlabel('Letras')
plt.ylabel('Valores')
plt.title('Ejemplo grafico de barras')
plt.show()
```

```python
x2 = ['A','B','C','D']
y2 = [100,50,80,40]
plt.pie(y2,labels=x2, autopct='%1.1f%%')#ver porcentajes
plt.title('Grafica de pie')
plt.axis('equal')#grafico circular y no eliptico
plt.show()
```

```python
x3 = [100,50,80,40]
y3 = [10,50,20,40]
z3 = [18,50,20,100]
fig = plt.figure()#lienzo en blanco
graf = fig.add_subplot(111,projection='3d')#dimensional
graf.plot(x3,y3,z3)
graf.set_xlabel('ejex')
graf.set_ylabel('ejey')
plt.title('Grafica 3D')
plt.show()
```

```python
from matplotlib import cm
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
x4 = [100,50,80,40]
y4 = [10,50,20,40]
#z4 = [18,50,20,100]
x4,y4 = np.meshgrid(x4,y4)
z4 = np.array([
    [18,50,20,100],
    [18,50,20,100],
    [18,50,20,100],
    [18,50,20,100]
])
fig = plt.figure()
graf = fig.add_subplot(111,projection='3d')
aux = graf.plot_surface(x4,y4,z4, cmap = cm.viridis)#Aplica un mapa de color (de morado a amarillo) basado en la altura (Z).
fig.colorbar(aux)#leyenda) al lado de la grfica, que muestra qu colores corresponden a qu valores de Z.
plt.title('Grafica 3D')
plt.show()
```
